# Proton-neutron scattering

The goal is to find phase shift and scattering cross sections for proton-neutron (pn) s-wave elastic scattering. To achieve this, i will use the Phase Function method that will be explained later on.


## Formulation of the problem

In the laboratory frame of reference we consider a proton at rest and a neutron with energy $E_{lab}$.

We will work in the center of mass reference system, where we look at a 'single particle' with position vector $\overrightarrow{r}$, reduced mass $μ_{cm}$ and energy $E$, such that

\begin{equation}\tag{1}
μ_{cm} = (\frac{m_pm_n}{m_n + m_p})
\end{equation}

\begin{equation}\tag{2}
E = (\frac{m_p}{m_n + m_p})E_{lab}
\end{equation}


---

To solve this problem we look at the time independant radial Schrödinger equation with $l=0$

$$
\frac{d^2u(r)}{dr^2}+\frac{2μ}{ħ^2}(E-V(r))u(r)=0
$$
where $u(r)= rR(r)$.

We also need to recall the solution for a particle scattering in a finite square well potential of width $r_0$ and depth $V_0$.

With the boundry conditions:

$$
u(0) = 0
$$

$$
u_{in}(r_0) = u_{out}(r_0)
$$

$$
\frac{du_{in}(r)}{dr}\bigg|_{r_0} = \frac{du_{out}(r)}{dr}\bigg|_{r_0}
$$

we find that

\begin{equation}\tag{3}
u_{in} = Asin(k_{in}r+ φ)
\end{equation}
\begin{equation}\tag{4}
u_{out} = Bsin(k_{out}r+ δ)
\end{equation}

here, $φ$ is a phase, $δ$ is the phase shift due to the scattering and

$$
k_{in} = \bigg(\frac{2μ(E-V_0)}{ħ^2}\bigg)^{\frac{1}{2}}
$$

$$
k_{out} = \bigg(\frac{2μE}{ħ^2}\bigg)^{\frac{1}{2}}
$$

### Phase function method

The goal here is to transform the Schrödinger equation into a differential equation for δ, the phase shift, that has the form

$$
\frac{dδ}{dr} = f(r,δ)
$$

To do this let us think of a general potential that consists of a combination of extremely small square wells of varying depths $V_i(r_i)$

<img src="https://drive.google.com/uc?export=view&id=1X0cfef-Ppy5IF40Yk4QkrAjP8EGcnNFW" width="500"/>

The $i^{th}$ well satisfies that

\begin{equation}\tag{5}
u_{i_{in}} = A_isin(k_{i_{in}}r+ φ_i)
\end{equation}
$$
k_{i_{in}} = \bigg(\frac{2μ(E-V_i(r_i))}{ħ^2}\bigg)^{\frac{1}{2}}
$$
With the same boundry conditions as before

\begin{equation}\tag{6}
u_{i_{in}}(r_{i+1}) = u_{i_{out}}(r_{i+1})
\end{equation}

\begin{equation}\tag{7}
\frac{du_{i_{in}}(r)}{dr}\bigg|_{r_{i+1}} = \frac{du_{i_{out}}(r)}{dr}\bigg|_{r_{i+1}}
\end{equation}

If we devide equation 7 by equation 6 we have

\begin{equation}\tag{8}
Z_i(r_{i+1}) = \frac{1}{u_{i_{in}}}\frac{du_{i_{in}}}{dr}\bigg|_{r_{i+1}} = k_{out}cot(k_{out}r_{i+1}+δ_i(r))
\end{equation}

Now, when the width of the wells tends to 0, the problem becomes continuous and we have that equations 5 and 8 are

\begin{equation}\tag{9}
u(r) = Asin(kr+ φ)
\end{equation}

$$
k = \bigg(\frac{2μ(E-V(r))}{ħ^2}\bigg)^{\frac{1}{2}}
$$

\begin{equation}\tag{10}
Z(r) = \frac{1}{u(r)}\frac{du(r)}{dr} = k_{out}cot(k_{out}r+δ(r))
\end{equation}

Now, taking the derivative of this equation we get

\begin{equation}\tag{11}
\frac{dZ(r)}{dr} = \frac{1}{u(r)}\frac{d^2u(r)}{dr^2}-Z(r)^2 = -\frac{k_{out}^2+k_{out}\frac{dδ}{dr}}{sin^2(k_{out}r+δ)}
\end{equation}

If we divide de radial Schrödinger equation by $u(r)$ and replace equations 10 and 11 in the result we will get

\begin{equation}\tag{12}
\frac{dδ(r)}{dr} = -\frac{sin^2(k_{out}r+δ)}{k_{out}}\frac{2μV}{ħ^2}
\end{equation}

with the condition that $δ(0)=0$

## The potentials



The neutron proton interaction is usually modeled with the Yukawa potential, which has the form

$$
V_Y = -V_A(\frac{e^{-\mu_A r}}{r})
$$

With $μ_A$ a screening parameter and $V_A$ the strenght of interaction.

However, experimentally, this potential is only good up to energies of $50MeV$. Then, a modification to the Yukawa potential that shows good results up to $350MeV$, is the Malfliet-Tjon potential:
$$
V_{MT}(r) = -V_A(\frac{e^{-\mu_Ar}}{r}) + V_R(\frac{e^{-\mu_Rr}}{r})
$$

Where the first terms corresponds to an attractive part and the second to a repulsive one.


---

For this problem it is important to consider the total angular momentum $J$ of the system, that for this case can take the values of 1 or 0. The parameters of the potentials have been adjusted according to experimental data, and for each $J$ they have different values. For instance, according to Awasthi et al. [1], the values for these parameters are

| Potential     | $J$ |$V_R (MeV)$  |$V_A(MeV)$  |$μ_A(fm^{-1})$|$μ_R(fm^{-1})$|
|---------------|---|----------------|--------|------------|------------|
| Yukawa        | 1 |   -     | 50.25  |    0.37    |     -      |
|               | 0 |   -     | 41.79  |    0.61    |     -      |
| Malfliet-Tjon | 1 | 9435.57 |2134.88 |    2.54    |    5.08    |
|               | 0 | 6806.60 |1522.42 |    2.42    |    4.84    |



```python
#Parameters for Yukawa with J = 1
μ1 =0.37
V1 = 50.25

#Parameters for Yukawa with J = 0
μ0 =0.61
V0 = 41.79

# Parameters for Malfliet-Tjon with J = 1
μ_A1 = 2.54
μ_R1 = 2*μ_A1
V_A1 = 2134.88
V_R1 = 9435.57

# Parameters for Malfliet-Tjon with J = 0
μ_A0 = 2.42
μ_R0 = 2*μ_A0
V_A0 = 1522.42
V_R0 = 6806.60
```

```python
def Vy(r, v_a, mu_a, v_r = 0, mu_r = 0):
  V = - v_a * np.exp( - mu_a * r ) / r
  return V

def Vmt(r, v_a, mu_a, v_r, mu_r):
  V = (- v_a * np.exp(- mu_a * r) + v_r * np.exp(- mu_r * r))/r
  return V
```

```python
r = np.linspace(0, 4, 200)
plt.figure(figsize=(8, 6))
plt.title('Potentials', size = 20)
plt.xlabel("$r$ $(fm)$", fontsize=15)
plt.ylabel("$V(r)$ $(MeV)$  ", fontsize=15)
plt.ylim(-160, 60)

plt.axhline(0, color='black', linestyle='-', linewidth=1, zorder=1)
plt.plot(r, Vy(r, V1, μ1), color='orangered', label = 'Yukawa, $J = 1$')
plt.plot(r, Vy(r, V0, μ0),linestyle = 'dashed', color='orangered', label = 'Yukawa, $J=0$')
plt.plot(r, Vmt(r, V_A1, μ_A1, V_R1, μ_R1), color = 'dodgerblue', label = "Malfliet_Tjon, $J = 1$")
plt.plot(r, Vmt(r, V_A0, μ_A0, V_R0, μ_R0),linestyle = 'dashed', color = 'dodgerblue', label = "Malfliet_Tjon, $J = 0$")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
```

## Finding the Phase shift

We will now solve equation 12 using the RK4 method.

\begin{equation}
\frac{dδ(r)}{dr} = -\frac{sin^2(k_{out}r+δ)}{k_{out}}\frac{2μV}{ħ^2}
\end{equation}

The following values for $ħ$ and $m_p$ are taken from the Committee on Data for Science and Technology (CODATA) [2]

```python
# value of c*hbar in MeVfm
ch_bar = 197.329

# mass of the proton in MeV/c^2
mp = 938.272

# mass of the neutron in MeV/c^2
mn = 939.565

# reduced mass
μ = mn * mp / (mn + mp)
```

```python
def rk4(f, sl, il, num_points, δ0, k, V, v_a, mu_a, v_r = 0, mu_r = 0):
    h = ((sl-il)/num_points)
    data = []
    for r in np.linspace(il,sl,num_points):
        data.append((r, δ0))

        k1 = f(r, δ0, k, V, v_a, mu_a, v_r, mu_r)
        k2 = f(r + h/2, δ0 + (h/2)*k1, k ,V, v_a, mu_a, v_r, mu_r)
        k3 = f(r + h/2, δ0 +(h/2)*k2, k, V, v_a, mu_a, v_r, mu_r)
        k4 = f(r + h, δ0 + h*k3, k, V, v_a, mu_a, v_r, mu_r)

        δ1 = δ0 + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
        δ0 = δ1
    return data
```

```python
def E(Elab):
  return mp * Elab / (mn + mp)

def k(E):
  return np.sqrt(2 * μ * E / ch_bar**2)

def delta_deriv(r, δ, k, V, v_a, mu_a, v_r, mu_r):
  return - (np.sin(k*r + δ)**2) * 2 * μ * V(r, v_a, mu_a, v_r, mu_r) /( k * ch_bar**2 )
```

```python
rmin = 0.08
rmax = 5
num_points = 500

# the chosen lab energy in MeV is 20
Ecm = E(20)
K = k(Ecm)
```

### Cross Section vs. Energy

From the partial wave analysis, we know the cross section can be calculated as

$$
σ(k)_{0,1} = \frac{4π}{k^2}sin^2δ_{1,0}
$$

where the phase shift value is the one the solution converges to after the potential.

Then, the total cross section is found as

$$
σ(k) = \frac{3}{4}σ_1 + \frac{1}{4}σ_0
$$



```python
def CrossSection(E, δ):
  return 4 * np.pi * np.sin(δ)**2 / k(E)**2

# phase shift for J = 1 with the MT potential
def delta1(E):
  return rk4(delta_deriv, rmax, rmin, num_points, 0, k(E), Vmt, V_A1, μ_A1, V_R1, μ_R1)[-1][-1]

# phase shift for J = 0 with the MT potential
def delta0(E):
  return rk4(delta_deriv, rmax, rmin, num_points, 0, k(E), Vmt, V_A0, μ_A0, V_R0, μ_R0)[-1][-1]

# total cross section
def TotalCrossSection(E):
  return 3/4 * CrossSection(E, delta1(E)) + 1/4 * CrossSection(E, delta0(E))
```
